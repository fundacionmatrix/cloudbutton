# FOREST FUEL MODELLING

## 1. Introduction
There are several approachments for modelling wildfire risk.

Forest fuel models are needed for several wildfire risk models such as BEHAVE. Its modelling is well studied through the years and it is based on information from a very wide range of disciplines. In this approachment, forest fuel of Cabañeros, a National Park of Spain, will be modelled from remote sensing images and LiDAR point clouds.

This notebook is part of a greater project (CloudButton) funded by the european initiative H2020 and it will test, among other projects, the performarance and the usability of CloudButton project.

```python
def find_LiDARgrid(las_tiles):
    def filestopath(path, lista):
        o = list(map(lambda x: f"{path}/{x}", lista))
        return(o)
        
    tmp = list(map(lambda x: f"{las_tiles}/{x}", os.listdir(las_tiles)))
    tmp2 = list(map(lambda x: filestopath(x, os.listdir(x)), tmp))
    tmp3 = sum(tmp2, [])
    out = list(filter(lambda x: x.endswith(".gpkg"), tmp3))
    return(out)
```
